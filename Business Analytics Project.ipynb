```python
# Task 1 - Data Overview

print(df.shape)
print(df.columns)
```

```python
# Task 2 - Gender Distribution

print("\n===== GENDER DISTRIBUTION =====")

# Replace 'gender' with your exact column name if needed
gender_count = df['gender'].value_counts()

print(gender_count)

gender_count.plot(kind='bar')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
# Task 3 - Descriptive Statistics

print("\n===== DESCRIPTIVE STATISTICS =====")

print(df.describe())

# Example for Age Column
if 'age' in df.columns:
    print("\nMean Age:", df['age'].mean())
    print("Median Age:", df['age'].median())
    print("Standard Deviation:", df['age'].std())
```

```python
# Task 4 - Most Preferred Investment

print("\n===== INVESTMENT AVENUE =====")

# Replace with exact column name
investment = df['Investment_Avenues'].value_counts()

print(investment)

investment.plot(kind='bar')

plt.title("Investment Preferences")
plt.xlabel("Investment Type")
plt.ylabel("Count")

plt.show()
```

```python
# Task 5 Reasons for Investment

print("\n===== REASONS FOR INVESTMENT =====")

# Replace with exact reason column
reason = df['Reason_Equity'].value_counts()

print(reason)
```

```python
# Task 6 - Savings Objectives

savings = df['What are your savings objectives?'].value_counts()

print("\nSavings Objectives:")
print(savings)

savings.plot(kind='pie', autopct='%1.1f%%')
plt.title("Savings Objectives")
plt.ylabel("")
plt.show()
```

```python
# Task 7 - Information Sources

print("\n===== INFORMATION SOURCES =====")

source = df['Source'].value_counts()

print(source)

source.plot(kind='bar')

plt.title("Information Sources")
plt.xlabel("Sources")
plt.ylabel("Count")

plt.show()
```

```python
# Task 8 - Investment Duration

print("\n===== INVESTMENT DURATION =====")

duration = df['Duration'].value_counts()

print(duration)
```

```python
# Task 9 - Expectations

print("\n===== EXPECTATIONS =====")

expect = df['Expect'].value_counts()

print(expect)
```

```python
# Task 10 - Correlation Analysis

print("\n===== CORRELATION ANALYSIS =====")

correlation = df.corr(numeric_only=True)

print(correlation)

sns.heatmap(correlation, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
